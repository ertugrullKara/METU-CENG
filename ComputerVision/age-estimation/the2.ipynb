```python
import matplotlib.pyplot as plt
%matplotlib inline

bs = 96
hidden_size = 3
best_net = None
best_val = np.inf
for lr in [1e-5]:
    for hidden_length in [[1024, 128], [1024], [1024, 64]]:
        net = Network(hidden_sizes=hidden_length, num_epochs=50000, batch_size=bs, learning_rate=lr)
        tr_loss_hist, val_loss_hist = net.train(early_stop_epochs=20, verbose=1)
        if min(val_loss_hist) < best_val:
            best_net = net
        plt.plot(tr_loss_hist[5:], label="train loss {} {} {} {}".format(lr, bs, hidden_size, hidden_length))
        plt.plot(val_loss_hist[5:], label="validation loss {} {} {} {}".format(lr, bs, hidden_size, hidden_length))

plt.xlabel("Epoch")
plt.ylabel("Loss")
lgd = plt.legend(loc=9, fancybox=True, shadow=True,
          bbox_to_anchor=(0.5, -0.2))
plt.savefig("fig{}.png".format("test"), bbox_extra_artists=(lgd,), bbox_inches="tight")
plt.show()
# plt.xticks(cluster_sizes)
```
